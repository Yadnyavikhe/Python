```python
#WHILE LOOP
i=1
while i<10:
    print(i,"Yadnya")
    i=i+1
```

```python
c=0
while c<11:
    print(c)
    c=c+1
```

```python
#REVERSE PATTERN
i=10
while i>=1:
    print(i,"Yadnya")
    i=i-1
    print(i)
```

```python
for i in range(1,11):
    print("2 *",i,"=",i*2)
```

```python
for i in range(1,11):
    print(2*i)
```

```python
#REVERSE RANGE PRINT

for i in range(10,0,-1):
    print(i)
```
