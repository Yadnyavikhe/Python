```python
import matplotlib.pyplot as plt
import matplotlib.style as mplstyle
mplstyle.use("ggplot")
plt.figure(figsize=(16,9))
classes = ["Python", "R", "AI", "ML", "DS"]
c1 = [30, 10, 20, 25, 10] # out of 100 student in each class
c2 = [40, 5, 20, 20, 10]
c3= [35, 5, 30, 15, 15] 
plt.bar(classes,c1,width=0.2,align="edge",color="k" ,edgecolor="g",linewidth=5
,alpha=0.5,linestyle="-",label="c1_student")
```

```python
mplstyle.use("ggplot")
plt.figure(figsize=(16,9))

plt.barh(classes,c1,align="edge",color="k" ,edgecolor="g",linewidth=5
,alpha=0.5,linestyle="-",label="c1_student")
```

```python
plt.title("Bar charts")
plt.xlabel("Classes")
plt.ylabel("no of students")

plt.bar(classes,c1,color="k" )

plt.bar(classes,c2,color="g")

plt.bar(classes,c3,color="m")
```

```python
import numpy as np
plt.title("Bar charts")
plt.xlabel("Classes")
plt.ylabel("no of students")
classes_index=np.arange(len(classes))

width=0.2
plt.xticks(classes_index + width , classes)
plt.bar(classes_index,c1,color="k" )

plt.bar(classes_index + width,c2,color="g")

plt.bar(classes_index+ width,c3,color="m")
```
