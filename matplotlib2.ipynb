```python
import matplotlib.pyplot as plt

# Define some sample data to use for the plot
classes = ['A', 'B', 'C', 'D','E']
c1 = [50,10,20,25,10]

plt.style.use("ggplot")
plt.figure(figsize=(16, 9))

plt.bar(
    classes, c1, width=0.2, align="edge", color="k", edgecolor="g", linewidth=5,
    alpha=0.5, linestyle="-", label="c1_student"
)

plt.title("Sample Bar Chart")
plt.xlabel("Classes")
plt.ylabel("Scores")
plt.legend()
plt.show()
```
